```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('/content/IOT_Devices_data.xlsx - Sheet1.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df["Error Code"] = df["Error Code"].fillna("unavailable")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
new_df = df.drop(columns = ["Timestamp", "id"])
new_df
```

```python
df_cat = new_df.drop(columns = ["Temperature (C)", "Power Consumption (W)"])
df_cat
```

```python
num_var = [key for key in dict(new_df.dtypes) if dict(new_df.dtypes)[key] in ["float64", "int64", "float32", "int32"]]
```

```python
num_var
```

```python
df_num = new_df[num_var]
df_num
```

```python
def create_dummies(df, col):
  col_d = pd.get_dummies(df[col], prefix = col, drop_first = True)
  df = pd.concat([df, col_d], axis = 1)
  df.drop(col, axis = 1, inplace = True)
  return df
```

```python
for c in ["Device ID", "Error Code", "Maintenance Required"]:
  df_cat = create_dummies(df_cat, c)
df_cat.head()
```

```python
df_cat.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output = False, drop = "first")
df_cat = pd.DataFrame(ohe.fit_transform(df_cat), columns = df_cat.columns)
df_cat
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
df3 = pd.DataFrame(ss.fit_transform(df_num), columns = df_num.columns)
```

```python
df3
```

```python
df_f = pd.concat([df3, df_cat], axis = 1)
df_f
```

```python
input_df = df_f.drop(columns="Maintenance Required_Yes")
output_df = df_f["Maintenance Required_Yes"]
input_df
```

```python
output_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_df, output_df, test_size=0.2, random_state = 42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100, lr.score(x_train, y_train)*100
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)*100
```
